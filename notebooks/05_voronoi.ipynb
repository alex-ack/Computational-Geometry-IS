```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
# Generating random points
def generate_points(num_points=10, seed=None):
    """Generates a set of random points in 2D space."""
    if seed:
        np.random.seed(seed)
    return np.random.rand(num_points, 2)

# Generate 10 points for this example
points = generate_points(10, seed=42)
points
```

```python
# Compute the Voronoi Diagram
def compute_voronoi(points):
    """Computes the Voronoi diagram for a given set of points."""
    return Voronoi(points)

# Compute Voronoi diagram for the points
vor = compute_voronoi(points)
```

```python
# Visualize the Voronoi diagram
def plot_voronoi(vor):
    """Plots the Voronoi diagram using Matplotlib."""
    voronoi_plot_2d(vor)
    plt.title("Voronoi Diagram")
    plt.show()

# Plot the diagram
plot_voronoi(vor)
```
