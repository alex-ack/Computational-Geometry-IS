```python
import matplotlib.pyplot as plt
from basics.point import Point
from algorithms.convex_hull import graham_scan

# Create a set of sample points
points = [Point(0, 0), Point(1, 1), Point(2, 2), Point(0, 2), Point(2, 0)]

# Get the convex hull points using the algorithm
hull = graham_scan(points)

# Plot the points
plt.scatter([p.x for p in points], [p.y for p in points], color='blue', label='Points')

# Plot the convex hull
hull_x = [p.x for p in hull] + [hull[0].x]
hull_y = [p.y for p in hull] + [hull[0].y]
plt.plot(hull_x, hull_y, 'r-', label='Convex Hull')

# Display the plot
plt.title("Convex Hull Visualization")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
