```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, Delaunay, voronoi_plot_2d

# Generate random points
def generate_points(num_points=10, seed=None):
    """Generates a set of random points in 2D space."""
    if seed is not None:
        np.random.seed(seed)
    return np.random.rand(num_points, 2)

# Generate 10 points with a seed
points = generate_points(10, seed=42)
points
```

```python
def compute_delaunay(points):
    return Delaunay(points)

# Compute Delaunay triangulation
delaunay = compute_delaunay(points)
delaunay
```

```python
def plot_delaunay(points, delaunay):
    """Plots the Delaunay triangulation."""
    plt.triplot(points[:, 0], points[:, 1], delaunay.simplices)
    plt.scatter(points[:, 0], points[:, 1], color='red')
    plt.title("Delaunay Triangulation")
    plt.show()

plot_delaunay(points, delaunay)
```

```python
def compute_voronoi(points):
    """Computes the Voronoi diagram for a set of points."""
    return Voronoi(points)

def plot_voronoi(vor):
    """Plots the Voronoi diagram using Matplotlib."""
    voronoi_plot_2d(vor)
    plt.title("Voronoi Diagram")
    plt.show()

vor = compute_voronoi(points)
plot_voronoi(vor)
```

```python
def plot_voronoi_delaunay(points, vor, delaunay):
    """Plots both Voronoi diagram and Delaunay triangulation together."""
    plt.triplot(points[:, 0], points[:, 1], delaunay.simplices, label="Delaunay")
    voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=1.5)
    plt.scatter(points[:, 0], points[:, 1], color='red')
    plt.title("Voronoi and Delaunay Diagram")
    plt.show()

# Plot both Voronoi and Delaunay
plot_voronoi_delaunay(points, vor, delaunay)
```
